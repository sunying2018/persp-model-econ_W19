5.1

If the individual lives for one period, the optimal problem is: 
$$\max \limits_{W_2\in[0,W_1]}u(W_1-W_2)$$
The optimal decision is characterized by:
$$W_2 = 0$$

5.2

The condition that characterizes the optimal amount of cake to leave for the next period  $W_3$  in period 2 is:

$$\max \limits_{W_3\in[0,W_2]}u(W_2-W_3)$$

The condition that characterizes the optimal amount of cake to leave for the next period  $W_2$  in period 1 is:

$$\max \limits_{W_2\in[0,W_1]}[u(W_1-W_2)+\max \limits_{W_3\in[0,W_2]}\beta u(W_2-W_3)]$$
The optimal decision is characterized by:
$$W_3 = 0$$
$$-u'(W_1-W_2)+\beta u'(W_2-W_3)=0$$


5.3

If the individual lives for three period, the conditions that characterizes the optimal amount of cake to leave for the next period
in period $W_2$:
$$\max \limits_{W_2\in[0,W_1]}\{u(W_1-W_2)+\max \limits_{W_3\in[0,W_2]}\beta [u(W_2-W_3)+\max \limits_{W_4\in[0,W_3]}\beta u(W_3-W_4)]\}$$
in period $W_3$:
$$\max \limits_{W_3\in[0,W_2]}\beta [u(W_2-W_3)+\max \limits_{W_4\in[0,W_3]}\beta u(W_3-W_4)]$$
in period $W_4$:
$$\max \limits_{W_4\in[0,W_3]} u(W_3-W_4)$$
The optimal decision is characterized by:
$$W_4=0$$
$$-u'(W_1-W_2)+\beta u'(W_2-W_3)=0$$
$$-\beta u'(W_2-W_3)+\beta^2 u'(W_3-W_4)=0$$
Assume that $u(x)=ln(x)$,$W_1=1$ and $W_4=0$
Then we can solve the equations to get the following consequences:
$$W_2=0.631$$
$$W_3=0.299$$ 
$$c_1=W_1-W_2=0.369$$
$$c_2=W_2-W_3=0.332$$
$$c_3=W_3-W_4=0.299$$
The evolution of $\{c_t\}_{t=1}^{3}$ and $\{W_t\}_{t=1}^{4}$ is as follows:

5.4

The condition that characterizes the optimal choice in period T-1 is:
$$-u'(W_{T-1}-\psi_{T-1}(W_{T-1}))+\beta u'(\psi_{T-1}(W_{T-1})) = 0$$
Then we can express $V_{T-1}$ as follows:
$$V_{T-1}(W_{T-1})=u(W_{T-1}-\psi_{T-1}(W_{T-1}))+\beta u(\psi_{T-1}(W_{T-1}))$$

5.5

According to 5.4,
$$\psi_{T}(\bar{W}) = W_{T+1}=0$$
$$\psi_{T-1}(\bar{W}) = W_{T}\neq 0$$
then we get get
$$\psi_{T} \neq \psi_{T-1} $$

$$V_T(\bar{W})=ln(\bar{W})$$
$$V_{T-1}(\bar{W})=ln(\frac{\bar{W}}{1+\beta})+\beta ln(\frac{\beta\bar{W}}{1+\beta})$$
So $$V_T(\bar{W}) \neq V_{T-1}(\bar{W})$$


5.6

The finite horizon Bellman equation for the value function at time T-2 is:
$$V_{T-2}(W_{T-2})=\max \limits_{W_{T-1}} ln(W_{T-2}-W_{T-1}))+\beta ln(\frac{W_{T-1}}{1+\beta})+\beta^2 ln(\frac{\beta W_{T-1}}{1+\beta})$$
The condition that characterizes the optimal choice in period T-2 is:
$$-\frac{1}{(W_{T-2}-\psi_{T-2}(W_{T-2}))}+(\beta+\beta^2)\frac{1}{\psi_{T-2}(W_{T-2})} = 0$$
The analytical solution for $\psi_{T-2}(W_{T-2})$ and $V_{T-2}(W_{T-2})$ is:
$$\psi_{T-2}(W_{T-2})=\frac{\beta+\beta^{2}}{1+\beta+\beta^{2}}W_{T-2}$$
$$V_{T-2}(W_{T-2}) = ln(\frac{W_{T-2}}{1+\beta+\beta^{2}})+\beta ln(\frac{\beta W_{T-2}}{1+\beta+\beta^{2}})+\beta^2 ln(\frac{\beta^2 W_{T-2}}{1+\beta+\beta^{2}})$$

5.7

Based on mathematical induction, the analytical solution for $\psi_{T-s}(W_{T-s})$ and $V_{T-s}(W_{T-s})$ is:
$$\psi_{T-s}(W_{T-s})=\frac{\sum\limits_{i=1}^{s}\beta^{i}}{1+\sum\limits_{i=1}^{s}\beta^{i}}W_{T-s}$$
$$V_{T-s}(W_{T-s}) = \left[\sum\limits_{i=0}^{s} \beta^{i}ln\left(\frac{\beta^{i}W_{T-s}}{1+\sum\limits_{i=1}^{s}\beta^{i}}\right)\right]$$

When the horizon becomes infinite:
$$\psi(W_{T-s})=\beta W_{T-s}$$
$$V(W_{T-s}) = \left(\frac{1}{1-\beta}\right)ln((1-\beta)W_{T-s})+\frac{\beta}{(1-\beta)^2}ln(\beta)$$

5.8

$$V(W)=\max \limits_{w\in[0,W]} u(W-w))+\beta V(w)$$

```python
import numpy as np

W_lb = 1e-2
W_ub = 1.0
N = 100
W_vec = np.linspace(W_lb, W_ub, N)
W_vec
```

5.10

```python
beta = 0.9
def utility(c):
    util = np.log(c)
    return util
V_init = np.zeros_like(W_vec)
W_prime = np.zeros_like(W_vec)
u_vec = utility(W_vec-W_prime)
V_T = (u_vec + beta * V_init)
print("The policy function is:\n", W_prime)
print("The value function is:\n", V_T)
```

5.11

```python
def distance(V_new, V_init):
    distance_metric = ((V_new-V_init)**2).sum()
    return distance_metric

delta1 = distance(V_T, V_init)
print("The distance metric of T is:", delta1)
```

```python
maxiters = 500
toler = 1e-9
dist = 10.0
VF_iter = 0
W_prime = W_vec
V_init = np.zeros_like(W_vec)
while dist >= toler and VF_iter < maxiters:
    VF_iter += 1
    W = W_prime
    V_prime = np.tile(V_init.reshape((1,N)), (N,1))
    V_prime[~c_pos] = -9e+4
    V_new = (u_mat + beta * V_prime).max(axis = 1)
    W_index = np.argmax(u_mat + beta*V_prime, axis=1)
    W_prime = W_vec[W_index]
    dist =((V_new - V_init)**2).sum()
    V_init = V_new
    print("Iteration=",VF_iter,",distance=", dist)
    
print("Converged.")
print("psi(W) is", W_prime)
print("V(W) is", V_init)
print("After {} iterations, V(W) converged.".format(VF_iter))
```

```python
from scipy.stats import norm

sigma = 0.5
mu = 4*sigma
M = 7
epsilon = np.linspace(mu - 3*sigma, mu + 3*sigma, M)
def gamma(x):
    gamma = norm(loc=mu, scale=sigma).pdf(x)
    return gamma
gamma = gamma(epsilon)
print("The epsilon is:\n", epsilon)
print("The probability distribution is:\n", gamma)
```

```python
plt.plot(epsilon,gamma)
plt.show()
```
